```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python
#Cargar archivo csv desde el equipo
from google.colab import files
files.upload()
```

```python
#¢Convertir en detaframe
data=pd.read_csv("PROCESADO_Recoleccion_de_datos_Negocios_IN1002B (1).csv")
data.head(5)
```

```python
#Verificamos información del DataFrame
data.info()
```

**Selección de variables categóricas (10)**

```python
#Filtro por columnas no consecutivas
categoricas=data.iloc[ : ,[4,9,10,11,13,14,15,18,20,21]]
categoricas
```

```python
#Verificamos información del nuevo DataFrame
categoricas.info()
```

```python
#Obtengo un análisis univariado de las variables categóricas
freq_tbl(categoricas)
```

Tabla 1

```python
#Obtengo un análisis univariado de una variable categórica en específico
table1= freq_tbl(categoricas['Grupo al cual pertenece'])
table1
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro= table1[table1['frequency']>1]
Filtro
```

```python
#Ajusto el indice de mi dataframe
Filtro_index= Filtro.set_index('Grupo al cual pertenece')
Filtro_index
```

```python
#Realizamos grafico de pastel del dataframe filtrado
Filtro_index["frequency"].plot(kind='pie', figsize=(10,5), shadow=False, autopct="%0.1f %%")
```
